```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
time_intervals = list(range(200, 3601, 200))
time_str = [str(i) for i in time_intervals]
```

## Random

```python
random_v = np.array([])
for i in range(1, 11):
    random = pd.read_csv('result2/r' + str(i) + '.csv')
    random_v = np.concatenate((random_v, random['oob'].values))

random_v = np.round(random_v, 2)
random_v = np.unique(random_v)
# top5
random_v = random_v[np.argsort(random_v)[-5:]]
display(random_v)
```

```python
new_values = []
for i in range(1, 11):
    random = pd.read_csv('result2/r' + str(i) + '.csv')
    value = [random[random['time'] < interval]['oob'].max() for interval in time_intervals]
    new_values.append(value)

df1 = pd.DataFrame(new_values, columns=time_str)

# 画 (2,1)的子图
fig, ax = plt.subplots(2, 1, figsize=(10, 6))
# plot1
ax[0].boxplot([df1[str(col)] for col in time_intervals], patch_artist=True,
              labels=time_str)
ax[0].set_title("Random Algorithm")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Fitness")

# plot2
mean1 = df1.copy().mean().to_frame()
ax[1].plot(mean1, color='blue', label='mean')
ax[1].set_title("Area Under Curve of Random Algorithm")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Fitness")
# fill the area under the curve with light blue
ax[1].fill_between(mean1.index, mean1[0], color='skyblue', alpha=0.5)

plt.tight_layout()
```
